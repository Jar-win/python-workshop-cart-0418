```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]
```

## Scipy Linear Model

```python
plt.scatter(X, Y)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(X,Y)
```

```python
slope
```

```python
intercept
```

$$y = mx + c$$

```python
lin = [intercept + slope * xi for xi in X] # calculating the of the line
```

```python
lin
```

```python
plt.scatter(X, Y, color="r")
plt.plot(X,lin, 'g--')
plt.show()
```

## Regression Using Seaborn

**Ascombe Dataset**

Anscombe's quartet comprises four datasets that have nearly identical simple descriptive statistics, yet appear very different when graphed illustrating the importance of looking at a set of data graphically before starting to analyze according to a particular type of relationship.

```python
import seaborn as sns
sns.set(style="ticks")
```

```python
df = sns.load_dataset("anscombe")
```

```python
df.head(n=3)
```

```python
df.tail(n=3)
```

```python

# Show the results of a linear regression within each dataset
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=df, ci=None, col_wrap=2)
```

**Polynomial Fitting**

```python
dataset_2 = df[df['dataset']=='II']
```

```python
sns.lmplot(x="x", y="y", data=dataset_2, order=2)
```

```python
iris = sns.load_dataset('iris')
```

```python
x1 = iris["sepal_length"]
x2 = iris["sepal_width"]
```

```python
g = sns.jointplot(x1, x2, kind="kde", size=7, space=0)
```

## KDE Plot

```python
setosa = iris[iris["species"] == 'setosa']
virginica = iris[iris["species"] == 'virginica']
```

```python
#styling of the plot
sns.set(style="darkgrid")
f, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")

# sepal_width vs sepal_length of setosa
sns.kdeplot(setosa.sepal_width, setosa.sepal_length,
                 cmap="Reds", shade=True, shade_lowest=False)

# sepal_width vs sepal_length of virginica
sns.kdeplot(virginica.sepal_width, virginica.sepal_length,
                 cmap="Blues", shade=True, shade_lowest=False)

ax.text(2.5, 8.2, "virginica", size=16)
ax.text(3.8, 4.5, "setosa", size=16)
```

Linear model

```python
sns.lmplot(x="sepal_length", y="sepal_width", hue="species", size=5, data=iris)
```
