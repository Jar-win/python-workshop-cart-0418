Source for csv: https://raw.githubusercontent.com/cogmaster-stats/python-cogstats/master/examples/brain_size.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('brain_size.csv', sep=';', na_values='.')
```

```python
# Box plots of different columns for each gender
groupby_gender = data.groupby('Gender')
groupby_gender.boxplot(column=['FSIQ', 'VIQ', 'PIQ'])
```

```python
# Scatter matrices for different columns
pd.tools.plotting.scatter_matrix(data[['Weight', 'Height', 'MRI_Count']])
pd.tools.plotting.scatter_matrix(data[['PIQ', 'VIQ', 'FSIQ']])

plt.show()
```
