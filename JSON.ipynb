```python
import json
```

```python
# moslty from web resources
f = open("landline_subscription.txt", 'r')
j = f.read()
```

```python
dictonary = json.loads(j) # convert json to dict
```

```python
dictonary[2]
```

```python
x = []
y = []
for d in dictonary:
    x.append(d['year'])
    y.append(d['value'])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("seaborn")
plt.plot(x, y, 'ro-')
plt.title("Landline Subscriptions in India")
plt.xlabel("Year")
plt.ylabel("Value")
plt.show()
```
