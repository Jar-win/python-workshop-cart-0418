```python
import numpy as np
import matplotlib.pyplot as plt
```

Plot:
1. $$y = x^3$$
2. $$y = \sqrt(x)$$
3. $$e^x+sin(x) \pi$$

## styling options

```python
x = np.linspace(0,2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y, color="r",  linewidth=2.5,linestyle="-")
```

```python
x = np.linspace(0,2*np.pi, 100)
y = np.sin(x)
x1 = np.linspace(0,2*np.pi, 100)
y1 = np.cos(x)
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```

Plot 2 & 3 in the same graph

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.style.use('ggplot')
sin = np.sin(x)
cos = np.cos(x)
tan = np.tanh(x)
plt.plot(x,sin, 'r', label="$sin(x)$")
plt.plot(x,cos, 'y', label="cos")
plt.plot(x,tan, 'm', label="tan")
plt.xlabel("x - values")
plt.ylabel("y -values")
plt.axis()
plt.grid()
plt.title("e^x")
plt.legend(loc="best")
plt.show()
```
