```python
import numpy as np
import matplotlib.pyplot as plt
```

`semilogx` -> `x` log scale, `y` normal scale. <br/>
`semilogy` -> `y` log scale, `x` normal scale. <br/>
`loglog`-> both `x` and `y` on log scale. <br/>
`np.logspace()` -> evenly spaced points. <br/>

```python
x = np.linspace(-50,15,200)
y = x**2
y1 = x
y2 = x**5
y3 = x**7
y4 = x**3
y5 = x**4
plt.plot(x,y1)
plt.plot(x,y, 'bo-')
plt.plot(x,y2, 'gs-')
plt.plot(x,y3, 'ro-')
plt.plot(x,y4, 'go-')
plt.plot(x,y5, 'rs-')
plt.axis([-15,15,-15,15])
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.grid(b=True, which='major', color='black', linestyle='--')
plt.axhline(linewidth=2, color='black')
plt.axvline(linewidth=2, color='black')
```

```python
x = np.logspace(-1,1,50) #log(10) 0.1 = -1 | 10^-1 = 0.1 | 10^1 = 10
y = x**2
y1 = x
y2 = x**5
y3 = x**7
y4 = x**3
y5 = x**4
plt.loglog(x,y1)
plt.loglog(x,y, 'bo-')
plt.loglog(x,y2, 'gs-')
plt.loglog(x,y3, 'ro-')
plt.loglog(x,y4, 'go-')
plt.loglog(x,y5, 'rs-')
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.grid(b=True, which='major', color='black', linestyle='--')
plt.axhline(linewidth=2, color='black')
plt.axvline(linewidth=2, color='black')
```
