```python

import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph() #create node object
```

```python
#adding nodes
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(['u','v'])
print("Nodes: ", G.nodes())
nx.draw(G)
```

```python
#adding edges
G.add_edge(1,2)
G.add_edge("u","v")
G.add_edges_from([(1,3),(1,4),(1,5),(1,6)]) #list containing tuple
G.add_edge("u","w")
print("Edges: ", G.edges())
nx.draw(G)
```

```python
#deleting nodes
G.remove_node(2)
G.nodes()
G.remove_nodes_from([4,5])
print(G.nodes())
nx.draw(G)
```

```python
#deleting edge
G.remove_edge(1,3)
print(G.edges())
G.remove_edges_from([(1,2),('u','w')])
nx.draw(G)
```

## Zachary's Karate Club

```python
K = nx.karate_club_graph()
print("Nodes: ",K.nodes())
print("Edges: ", K.edges())
nx.draw(K, with_labels=True, node_color='lightblue', edge_color='grey')
```

    But this layout is completely counter intutive. Let's draw a circular graph to get a better overview for the above relationship
    
  instead of `nx.draw` use `nx.draw_circular`

```python
nx.draw_circular(K, with_labels=True, node_color='lightblue', edge_color='grey')
```
